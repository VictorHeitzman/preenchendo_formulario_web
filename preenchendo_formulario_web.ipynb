```python
import pandas as pd
from selenium import webdriver as driver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from time import sleep
```

```python
df = pd.read_excel('challenge.xlsx', header=0, 
                   names=['first_name',
                          'last_name',
                          'company_name',
                          'role_in_company',
                          'address',
                          'email',
                          'phone_number'])

df.head()
```

```python
driver = driver.Chrome()
driver.get('https://rpachallenge.com/')
```

```python
for i in df.itertuples(index=None):
    first_name = i.first_name
    last_name = i.last_name
    company_name = i.company_name
    role_in_company = i.role_in_company
    phone_number = i.phone_number
    address = i.address
    email = i.email

    driver.find_element(By.XPATH,'//*[@ng-reflect-name="labelFirstName"]').send_keys(first_name)
    driver.find_element(By.XPATH,'//*[@ng-reflect-name="labelRole"]').send_keys(role_in_company)
    driver.find_element(By.XPATH,'//*[@ng-reflect-name="labelLastName"]').send_keys(last_name)
    driver.find_element(By.XPATH,'//*[@ng-reflect-name="labelCompanyName"]').send_keys(company_name)
    driver.find_element(By.XPATH,'//*[@ng-reflect-name="labelPhone"]').send_keys(phone_number)
    driver.find_element(By.XPATH,'//*[@ng-reflect-name="labelEmail"]').send_keys(email)
    driver.find_element(By.XPATH,'//*[@ng-reflect-name="labelAddress"]').send_keys(address)

    sleep(0.5)
    driver.find_element(By.XPATH,'/html/body/app-root/div[2]/app-rpa1/div/div[2]/form/input').click()
    sleep(1)
    
```
